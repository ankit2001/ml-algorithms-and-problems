```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(19.0,16.0)
```

```python
data=pd.read_csv('/home/ankittt2001/Desktop/ml algorithms and problems/datasetse/train.csv')
```

```python
X=data.iloc[:,0]
Y=data.iloc[:,1]
plt.scatter(X,Y)
plt.show()
data.head()
```

```python
q1=0
q2=0
alpha=0.0001
n=float(len(X))
for i in range(100):
    Y_obs=q1*X+q2
    diff1=(-2/n)*sum(Y)
    diff2=(-2/n)*sum((Y-Y_obs)*X)
    q1=q1-diff2*alpha
    q2=q2-diff1*alpha
```
