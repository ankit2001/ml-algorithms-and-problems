```python
#splitting data in random values
def load(file,split,train,test):
    with open(file,'r') as f:
        lines=csv.reader(f)
        dataset=list(lines)
        print(dataset)
        for x in range(len(dataset)-1):
            for y in range(4):
                dataset[x][y]=float(dataset[x][y])
            if random.random()<split:
                train.append(dataset[x])
            else:    
                test.append(dataset[x])
```

```python
#euclidean distance
def distance(i,k,lent):
    distance=0
    for x in range(lent):
        distance+=pow(i[x]-k[x],2)
    return math.sqrt(distance)    
```

```python
import operator
```

```python
def closest_neighbour(train,test,k):
    all_dist=[]
    for x in range(len(train)):
        dist=distance(train[x],test,len(test)-1)
        all_dist.append((train[x],dist))
    all_dist.sort(key=operator.itemgetter(1))    
    print(all_dist)
    k_close_neg=[]
    for x in range(k):
        k_close_neg.append(all_dist[x])
    return k_close_neg
#just testing the function ,let randomly take value of k as 4

closest_neighbour(train,test[0],4)
```
